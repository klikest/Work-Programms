```python
import matplotlib.pyplot as plt
import math
```

```python
t = [0, 60, 120, 150, 170, 180, 182, 183, 185, 190, 200, 300]
Re_st = [12, 11.9, 11.3, 11,10,8,6,4,2,1.5,1,0.5]
Im_st = [0,-2,-3,-4,-5,-5.9,-6,-5.6,-4,-3.2,-2.8,-2]


Re_rez = [0.02,0.04,0.055,0.06,0.05,0.02,0,-0.02,-0.04,-0.045,-0.04,-0.03]
Im_rez = [0,-0.02,-0.04,-0.06,-0.1,-0.12,-0.125,-0.12,-0.1,-0.07,-0.04,-0.015]
```

```python
def plot_polar(fi, A):
    ax= plt.subplot(111, projection='polar')
    ax.set_theta_direction(-1)
    ax.plot(fi, A)
    plt.show()
    
```

```python
fi_st = []
A_st = []

fi_rez = []
A_rez = []

for i in range(len(Re_st)):
    A_st.append( (Im_st[i]**2 + Re_st[i]**2 )**(1/2) )
    fi_st.append(-math.atan2(Im_st[i], Re_st[i]))

for i in range(len(Re_rez)):
    A_rez.append( (Re_rez[i]**2 + Im_rez[i]**2 )**(1/2) )
    fi_rez.append(-math.atan2(Im_rez[i], Re_rez[i]))


print(fi_rez)
```

```python
plt.plot(Re_st, Im_st, label = "st")
plt.grid()
```

```python
ax= plt.subplot(111, projection='polar')
ax.set_theta_direction(-1)
ax.plot(fi_st, A_st)
plt.show()
```

```python
A_raz = []
fi_raz = []

for i in range(len(A_rez)):
    A_raz.append(A_rez[i]*A_st[i])
    fi_raz.append(fi_rez[i]+fi_st[i])
```

```python
ax= plt.subplot(111, projection='polar')
ax.set_theta_direction(-1)
ax.plot(fi_raz, A_raz)
ax.scatter(math.pi , 1, color = "red")
plt.show()
```

```python
Re_raz = []
Im_raz = []

for i in range(len(A_raz)):
    Re_raz.append(A_raz[i]*math.cos(-fi_raz[i]))
    Im_raz.append(A_raz[i]*math.sin(-fi_raz[i]))
```

```python
Re_zn = []
Im_zn = []
A_zn = []
fi_zn = []

for i in range(len(Re_raz)):
    Re_zn.append(Re_raz[i] + 1)
    Im_zn.append(Im_raz[i])  
    A_zn.append( (Re_zn[i]**2 + Im_zn[i]**2 )**(1/2) )
    fi_zn.append(-math.atan2(Im_zn[i], Re_zn[i]))
```

```python
ax= plt.subplot(111, projection='polar')
ax.set_theta_direction(-1)
ax.plot(fi_raz, A_raz)
ax.plot(fi_zn, A_zn)
ax.scatter(math.pi , 1, color = "red")
plt.show()
```

```python
Re_y = []
Im_y = []
A_y = []
fi_y = []

for i in range(len(A_raz)):
    A_y.append(A_raz[i]/A_zn[i])
    fi_y.append(fi_raz[i] - fi_zn[i])
```

```python
ax= plt.subplot(111, projection='polar')
ax.set_theta_direction(-1)
ax.plot(fi_raz, A_raz, label="$W_{ }$")
ax.plot(fi_zn, A_zn, label="$W_{}$")
ax.plot(fi_y, A_y, label='$W_{y}$')
ax.scatter(fi_y, A_y, c ="green", 
            linewidths = 1, 
            marker ="s", 
            edgecolor ="green", 
            s = 15)
ax.scatter(math.pi , 1, color = "red")
ax.legend()
plt.show()
plt.savefig('better.png', dpi=300)
```
