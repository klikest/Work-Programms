```python
import matplotlib.pyplot as plt
import math
PI = math.pi
```

```python
# Исходные данные

P = 956      # кВт
n_in = 14000 # об/мин
n_out = 1000 # об/мин
t_h = 2600   # ч
```

```python
# Материал зучатых колес

# Легированная сталь 12Х2Н4А с термообработкой цементацией.

HRC = 60
sigma_H_lim_b = 23*HRC
N_H0 = 30*(10*HRC)**2.4
N_H0 = 12*10**7
K_H = K_HLg = K_HLf = K_HLb = K_HL = 1

sigma_F_lim_b = 800
```

```python
slice = "----------------------------"
table_1_i_pl = [8, 10, 12, 14, 16]
table_1_K_r = [0.6, 1, 1.4, 1.8, 2.2]
acc = 0.1

def calculate_hcf(x, y): # selecting the smaller number 
    hcf = 0
    if x > y: 
        smaller = y 
    else: 
        smaller = x 
    for i in range(2,smaller + 1): 
        if((x % i == 0) and(y % i == 0)): 
            hcf = i 
    return hcf
```

```python
print("1. Кинематический и энергетический расчёт редуктора")
print(slice)
i_pl = n_in/n_out
print("Передаточное отношение редуктора i_pl =", i_pl)
err = []
for i in range(len(table_1_i_pl)):
    err.append(abs(i_pl-table_1_i_pl[i]))

temp = min(err)
res = [i for i, j in enumerate(err) if j == temp]
K_r = table_1_K_r[res[0]]
print(slice)
print("Выбрано отношение диаметров Kr =", K_r)

i_ag = (i_pl/(1+K_r))-1
i_fb = (i_pl-1)/i_ag
print(slice)
print("Передаточное отношение передачи i_ag =", i_ag)
print("Передаточное отношение передачи i_fb =", i_fb)

n_ah = n_in-n_out
n_gh = n_fh =n_ah/i_ag
n_bh = n_out
print(slice)
print("Частота вращения солнечного колеса =", n_ah, "об/мин")
print("Частота вращения сателлита =", n_gh, "об/мин")
print("Частота вращения водила =", n_bh, "об/мин")

a_c = (0.9*PI)/(math.asin( (i_pl-1-K_r) / (i_pl) ))
a_c = math.floor(a_c)
print(slice)
print("Максимальное количество сателитов a_c =", a_c)

P_a = P/0.9815
print(slice)
print("Мощность на входном валу P_a =", round(P_a, 1), "кВт")

T_a = 9.55*10**6 * P_a/n_in
print(slice)
print("Крутящий момент на входном валу T_a =", round(T_a, 1), "Н*мм")

K_ner =1
print(slice)
print("Коэффициент неравномерности распределения нагрузки между сателлитами K_нер = 1" )

T_ag = (T_a*K_ner)/a_c
print(slice)
print("Крутящий момент для ступени a-g, Tag =", round(T_ag, 1), "Н*мм" )

T_fb = T_ag*i_ag*0.99
print(slice)
print("Крутящий момент для ступени f-b, Tfb =", round(T_fb, 1), "Н*мм" )
print(" ")
print(" ")
print(" ")


print("2. Определение допускаемых напряжений")

S_H = 1.2
sigma_H = sigma_H_lim_b*K_HL/S_H
print(slice)
print("Принимаем запас прочности по контактным напряжениям S_H =", S_H)
print("Допускаемые контактные напряжения для колёс sigma_H =", sigma_H, "МПа")

S_F = 2
sigma_F = sigma_F_lim_b/S_F
print(slice)
print("Запас прочности по изгибным напряжениям примем равным S_F =", S_F )
print("Допускаемые напряжения изгиба для центрального, корончатого колёс и сателлитов", sigma_F, "МПа")
print(" ")
print(" ")
print(" ")




print("3. Проектировочный расчёт передачи")

K_H = 1.3
psi_bd = 0.6
print(slice)
print("Коэффициент нагрузки принимаем K_H =", K_H)
print("Коэффициент ширины зуба колеса принимаем psi_bd =", psi_bd)


d_a = 77*((T_ag*K_H*(i_ag+1)) / (psi_bd*sigma_H**2 *i_ag))**(1/3)
print(slice)
print("Диаметр центрального колеса из условия контактной прочности:", round(d_a), "мм")

b_1 = psi_bd*d_a
print(slice)
print("Ширина зуба центрального колеса и сателлита g", round(b_1), "мм")

Y_F = 4.1
K_F = 1.2
print(slice)
print("Коэффициент нагрузки принимаем равным К_F =", K_F)
print("Коэффициент формы зуба Y_F =", Y_F)

m_1 = (2*T_ag*K_F*Y_F) / (d_a*b_1*sigma_F)
print(slice)
print("Модуль передачи из условия изгибной прочности m =", round(m_1, 2), "мм")

modul_mass = [2.5, 2.75, 3, 3.5, 4, 4.5, 5, 5.5, 6, 7, 8, 9, 10, 11, 12, 14, 16, 18, 20, 22]
m1 = 0
for i in range(len(modul_mass)):
    if m_1 < modul_mass[i] or abs(m_1-modul_mass[i])/m_1 < 0.05:
        m1 = modul_mass[i]
        break

print("Значение модуля после округления до стандартного", round(m1, 2), "мм")

a_w = (d_a*(i_ag+1))/2
print(slice)
print("Ориентировочное значение межосевого расстояния a_w =", round(a_w, 1), "мм")

d_f = (2*a_w)/(i_fb-1)
print(slice)
print("Диаметр сателлита будет равен d_f", round(d_f, 2), "мм")


b_2 = 77**3 * (T_fb*K_H*(i_fb-1))/(sigma_H**2 * d_f**2 * i_fb)
print(slice)
print("Ширина сателлита f и корончатого колеса из условия контактной прочности:", round(b_2, 2), "мм")
b_2 = math.ceil(b_2)
print("Значение ширины после округления:", b_2, "мм")

m_2 = (2*T_fb*K_F*Y_F)/(d_f*b_2*sigma_F)
print(slice)
print("Модуль передачи из условия изгибной прочности:", round(m_2, 2),"мм")
m2 = 0
for i in range(len(modul_mass)):
    if m_2 < modul_mass[i] or abs(m_2-modul_mass[i])/m_2 < 0.1:
        m2 = modul_mass[i]
        break
print("Значение модуля после округления до стандартного", round(m2, 2), "мм")
print(" ")
print(" ")
print(" ")



print("4. Подбор чисел зубьев")


z_a_0 = round(d_a/m1, 2)
z_a = round(z_a_0)

z_g = round(z_a*i_ag)
z_f = round(d_f/m2)
z_b = round(z_f*i_fb)

k = calculate_hcf(z_g, z_f)
N = 1.5
while k == 0 and N == round(N):
    z_a += 1
    z_g = round(z_a*i_ag)
    z_f = round(d_f/m2)
    z_b = round(z_f*i_fb)
    k = calculate_hcf(z_g, z_f)
    if k != 0:
        N = (z_a*z_f + z_g*z_b)/(a_c*k)


print(slice)
print("Число зубьев солнечного колеса z_a=", z_a_0)
print("Принимаем z_a=", z_a)

print("z_g =", z_g)
print("z_f =", z_f)
print("z_b =", z_b)

k = calculate_hcf(z_g, z_f)

N = (z_a*z_f + z_g*z_b)/(a_c*k)
print("N =", N)
print("N - целое, следовательно условие сборки выполняется!")


```
